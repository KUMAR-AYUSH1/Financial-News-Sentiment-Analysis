```python
df1.rename(columns={'Sentiment':'setiment','Title':'text'},inplace=True)
df2.rename(columns={'Full_text':'text','Final Status':'setiment'},inplace=True)
```

```python
df = pd.concat([df,df1,df2],axis=0)
df.head()
```

```python
df.setiment.value_counts()
```

```python
df['setiment'] = df['setiment'].str.replace(' ','')
```

```python
df['setiment'] = df['setiment'].str.replace('Positive','positive')
df['setiment'] = df['setiment'].str.replace('Negative','negative')
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

handling imbalanced data

```python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=42)
x = df[['text']]      
y = df['setiment'] 
X_resampled, y_resampled = rus.fit_resample(x, y)

df = pd.concat([X_resampled,y_resampled],axis=1)
df = pd.DataFrame(df,columns=['text','setiment'])
df.head()
```

```python
df['setiment'].value_counts()
```

```python
df.to_csv('final_data.csv',index=False)
```
