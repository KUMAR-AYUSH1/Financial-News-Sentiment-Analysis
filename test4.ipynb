```python
import requests
from bs4 import BeautifulSoup 
```

```python
url ="https://www.moneycontrol.com/news/business/markets/sensex-today-stock-market-live-updates-nifty50-share-price-crude-fii-gift-nifty-rupee-latest-updates-30-04-2026-liveblog-13903914.html"
```

```python
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
print(soup)
```

```python
# Find the div using class_
container = soup.find('div', class_="jsx-b1c2aacb25c7150f disBdy").get_text(strip=True)

# Print the extracted text
print(container)
```
